```python
from pyspark.sql.functions import*
df_train = spark.table("irctc_catalog.silver.train_silver")

df_route_analytics = (
    df_train
    .groupBy("route")
    .agg(
        countDistinct("trainNumber").alias("train_count")
    )
)

display(df_route_analytics)
```

```python
df_route_analytics.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("irctc_catalog.gold.route_analytics_gold")
```

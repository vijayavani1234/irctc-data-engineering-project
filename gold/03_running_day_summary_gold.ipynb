```python
from pyspark.sql.functions import *

df_train = spark.table("irctc_catalog.silver.train_silver")

day_summary = df_train.selectExpr(
    "sum(cast(MON as int)) as MON",
    "sum(cast(TUE as int)) as TUE",
    "sum(cast(WED as int)) as WED",
    "sum(cast(THU as int)) as THU",
    "sum(cast(FRI as int)) as FRI",
    "sum(cast(SAT as int)) as SAT",
    "sum(cast(SUN as int)) as SUN"
)

display(day_summary)
```

```python
day_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("irctc_catalog.gold.running_day_summary_gold")
```

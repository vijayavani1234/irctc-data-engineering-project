```python
df_train = spark.table("irctc_catalog.silver.train_silver")

df_route = spark.table("irctc_catalog.silver.train_route_silver")
```

```python
print(df_train.count())
print(df_route.count())
```

```python
from pyspark.sql.functions import *

df_train_summary = df_route.groupBy("trainNumber", "trainName", "route").agg(
    count("station_seq").alias("total_stations")
)
```

```python
display(df_train_summary)
```

```python
df_train_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("irctc_catalog.gold.train_summary_gold")
```
